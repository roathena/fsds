```python
#q1

def fibo(num):
    if num ==0:
        return 0
    elif num == 1 or num == 2:
        return 1
    else:
        return fibo(num -1) + fibo(num-2)

for i in range(10000):
    print(fibo(i))
```

```python
#q2
def fact(num):
    if num==0:
        return 1
    else:
        return num*fact(num-1)
fact(5)
```

```python
def bmi(w, h):
    bmi = w/h
    if bmi < 18.5:
        return 'under weight'
    elif bmi >= 18.5 or bmi <= 24.9:
        return 'normal weight'
    else:
        return 'over weight'

bmi(87, 1.82)
```
